```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline

# evenly sampled time
t = np.arange(0.01, 500., 0.002)

mpl.rc('text', usetex = True)
mpl.rc('font', family = 'serif', size="16")

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10, 5))

# surface width, k = 0.1
ax1.loglog(t, np.sqrt(100*np.exp(-0.1*t)*(1.-np.exp(-0.1*t))), linewidth=2.5, color="green")
ax1.loglog(t, np.sqrt(1000*np.exp(-0.1*t)*(1.-np.exp(-0.1*t))), linewidth=2.5, color="green")
ax1.loglog(t, np.sqrt(10000*np.exp(-0.1*t)*(1.-np.exp(-0.1*t))), linewidth=2.5, color="green")

# surface width, k = 1
ax1.loglog(t, np.sqrt(100*np.exp(-t)*(1.-np.exp(-t))), linewidth=2.5, color="black")
ax1.loglog(t, np.sqrt(1000*np.exp(-t)*(1.-np.exp(-t))), linewidth=2.5, color="black")
ax1.loglog(t, np.sqrt(10000*np.exp(-t)*(1.-np.exp(-t))), linewidth=2.5, color="black")

# surface width, k = 10
ax1.loglog(t, np.sqrt(100*np.exp(-10*t)*(1.-np.exp(-10*t))), linewidth=2.5, color="#005A8B")
ax1.loglog(t, np.sqrt(1000*np.exp(-10*t)*(1.-np.exp(-10*t))), linewidth=2.5, color="#005A8B")
ax1.loglog(t, np.sqrt(10000*np.exp(-10*t)*(1.-np.exp(-10*t))), linewidth=2.5, color="#005A8B")

ax1.set_xscale("log", nonposx='clip')
ax1.set_yscale("log", nonposy='clip')

ax1.set_ylabel("$w(N,t)$")
ax1.set_xlabel("$t$")
ax1.text(0.09, 0.01, "$\\frac{\\ln 2}{k}$", size="22")

# vertical line at maximum: k^(-1) * ln 2
ax1.axvline(np.log(2.)/(0.1), color="black", linestyle="--")
ax1.axvline(np.log(2.)/(1.0), color="black", linestyle="--")
ax1.axvline(np.log(2.)/(10.), color="black", linestyle="--")

ax1.loglog(t, np.sqrt(100*t), linewidth=1.5, linestyle='dashed', color="red")
ax1.loglog(t, np.sqrt(np.exp(-10*t)), linewidth=1.5, linestyle='dashed', color="red")
ax1.loglog(t, np.sqrt(np.exp(-1*t)), linewidth=1.5, linestyle='dashed', color="red")
ax1.loglog(t, np.sqrt(np.exp(-0.1*t)), linewidth=1.5, linestyle='dashed', color="red")

ax1.set_xlim(-.1, 500)
ax1.set_ylim(0.0005, 100)

# data collapse

# surface width, k = 0.1
ax2.loglog(0.1*t, np.sqrt(100*np.exp(-0.1*t)*(1.-np.exp(-0.1*t)))/np.sqrt(100), linewidth=2.5, color="green")
ax2.loglog(0.1*t, np.sqrt(1000*np.exp(-0.1*t)*(1.-np.exp(-0.1*t)))/np.sqrt(1000), linewidth=2.5, color="green")
ax2.loglog(0.1*t, np.sqrt(10000*np.exp(-0.1*t)*(1.-np.exp(-0.1*t)))/np.sqrt(10000), linewidth=2.5, color="green")

# surface width, k = 1
ax2.loglog(t, np.sqrt(100*np.exp(-t)*(1.-np.exp(-t)))/np.sqrt(100), linewidth=2.5, color="black")
ax2.loglog(t, np.sqrt(1000*np.exp(-t)*(1.-np.exp(-t)))/np.sqrt(1000), linewidth=2.5, color="black")
ax2.loglog(t, np.sqrt(10000*np.exp(-t)*(1.-np.exp(-t)))/np.sqrt(10000), linewidth=2.5, color="black")

# surface width, k = 10
ax2.loglog(10*t, np.sqrt(100*np.exp(-10*t)*(1.-np.exp(-10*t)))/np.sqrt(100), linewidth=2.5, color="#005A8B")
ax2.loglog(10*t, np.sqrt(1000*np.exp(-10*t)*(1.-np.exp(-10*t)))/np.sqrt(1000), linewidth=2.5, color="#005A8B")
ax2.loglog(10*t, np.sqrt(10000*np.exp(-10*t)*(1.-np.exp(-10*t)))/np.sqrt(10000), linewidth=2.5, color="#005A8B")

ax2.loglog(t, np.sqrt(t), linewidth=1.5, linestyle='dashed', color="red")
ax2.loglog(t, np.sqrt(np.exp(-t)), linewidth=1.5, linestyle='dashed', color="red")
ax2.text(2, 0.5, "$\\sqrt{e^{-t}}$", color="red")
ax2.text(0.05, 0.5, "$t^{1/2}$", color="red")

ax2.set_xscale("log", nonposx='clip')
ax2.set_yscale("log", nonposy='clip')

ax2.set_ylabel("$w(N,t)/N^{\\alpha}$")
ax2.set_xlabel("$k t$")
ax2.text(1, 0.1, "$\\ln 2$")

# vertical line at maximum: ln 2
ax2.axvline(np.log(2.)/(1.0), color="black", linestyle="--")

ax2.set_xlim(-.1, 500)
ax2.set_ylim(0.0005, 1)

plt.tight_layout()
#plt.show()
plt.savefig("width-figure.pdf", bbox_inches='tight')
```
